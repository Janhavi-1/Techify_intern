### **1. Data Collection**
Obtain sales data over a period of time (daily, weekly, monthly).

```python
#reading the data
import pandas as pd
# Load the sales data

hourly = pd.read_csv('saleshourly.csv')
daily = pd.read_csv('salesdaily.csv')
weekly = pd.read_csv('salesweekly.csv')
monthly = pd.read_csv('salesmonthly.csv')
```

### **2.Data Cleaning**
 Clean and preprocess the data.

```python
# Check for missing values
print(weekly.isnull().sum())
print(monthly.isnull().sum())
print(hourly.isnull().sum())
print(daily.isnull().sum())
```

```python
# Convert date column to datetime type
monthly['datum'] = pd.to_datetime(monthly['datum'], format= '%Y-%m-%d')
weekly['datum'] = pd.to_datetime(weekly['datum'], format= '%m/%d/%Y')
daily['datum'] = pd.to_datetime(daily['datum'], format= '%m/%d/%Y')


# Sort data by date
monthly = monthly.sort_values(by='datum')
weekly = weekly.sort_values(by='datum')
daily = daily.sort_values(by='datum')

print(monthly.head())
print(weekly.head())
print(daily.head())

```

Analysing Monthly Series Firstly, let us analyse the monthly data and see what inferences can we draw from this data.

Extracting year, month and day from the datum column

```python
#extracting year from dates
monthly['year'] = monthly['datum'].dt.year
monthly['month'] = monthly['datum'].dt.month
monthly['day'] = monthly['datum'].dt.day
```

```python
monthly.set_index(monthly['datum'], inplace= True)
```

```python
monthly.sample(10)
```

```python
monthly.columns
```

```python
num_col_list=['M01AB', 'M01AE', 'N02BA', 'N02BE', 'N05B', 'N05C', 'R03',
       'R06']
```

### **3.	Trend Analysis**
 Analyze trends in sales data using time series analysis techniques.

```python
import matplotlib.pyplot as plt
for column in num_col_list:
    plt.figure(figsize=(10, 6))
    plt.plot(monthly.index, monthly[column])
    plt.title(f'{column} Time Series')
    plt.xlabel(column)
    plt.ylabel('Sales Value')
    plt.show()
```

```python
#calculating total sales
monthly['total_sales'] = monthly['M01AB']
for cols in monthly.columns[2:9]:
    monthly['total_sales'] = monthly['total_sales']+monthly[cols]
```

```python
# plotting the total sales data
plt.figure(figsize=(10, 6))
plt.plot(monthly.index, monthly['total_sales'])
plt.xlabel('Date Time')
plt.ylabel('Total Sales')
plt.title('Total Sales of Drugs')
plt.show()
```

### **4.	Seasonality**
 Identify any seasonal patterns in the sales data.

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Perform seasonal decomposition
#In this case, the period is 12 (for 12 months in a year).
decomposition = seasonal_decompose(monthly['total_sales'], model='additive', period=12)
decomposition.plot()
plt.show()
```

### **5. Forecasting**
For forecasting future sales, we'll use the ARIMA model.

```python
from statsmodels.tsa.arima.model import ARIMA

# Fit the ARIMA model
model = ARIMA(monthly['total_sales'], order=(5, 1, 0))  # Adjust p, d, q based on model selection
model_fit = model.fit()

# Forecast future sales
forecast = model_fit.forecast(steps=12)  # Forecasting the next 12 months
forecast_index = pd.date_range(start=monthly['datum'].iloc[-1] + pd.DateOffset(months=1), periods=12, freq='M')



# Plot the forecast
plt.figure(figsize=(10, 6))
plt.plot(monthly.index, monthly['total_sales'], label='Sales')
plt.plot(forecast_index, forecast, label='Forecast', color='red')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('Sales Forecast')
plt.legend()
plt.show()
```

### **6.	Visualization**
 Create visualizations to represent sales trends and forecasts.

```python
import seaborn as sns

# Create a heatmap for seasonal patterns
pivot_table = monthly.pivot_table(values='total_sales', index='year', columns='month', aggfunc='sum')

plt.figure(figsize=(12, 6))
sns.heatmap(pivot_table, annot=True, fmt='.0f', cmap='coolwarm')
plt.title('Sales Heatmap by Year and Month')
plt.xlabel('Month')
plt.ylabel('Year')
plt.show()
```
